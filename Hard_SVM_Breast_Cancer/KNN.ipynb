```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split
from collections import Counter
```

```python
cols = ['ID', 'diagnosis', "mean_radius",
    "mean_texture",
    "mean_perimeter",
    "mean_area",
    "mean_smoothness",
    "mean_compactness",
    "mean_concavity",
    "mean_concave_points",
    "mean_symmetry",
    "mean_fractal_dimension",
    "radius_se",
    "texture_se",
    "perimeter_se",
    "area_se",
    "smoothness_se",
    "compactness_se",
    "concavity_se",
    "concave_points_se",
    "symmetry_se",
    "fractal_dimension_se",
    "worst_radius",
    "worst_texture",
    "worst_perimeter",
    "worst_area",
    "worst_smoothness",
    "worst_compactness",
    "worst_concavity",
    "worst_concave_points",
    "worst_symmetry",
    "worst_fractal_dimension"]
df = pd.read_csv('wdbc_data.csv', names= cols)
df.head()
```

```python
df = df.drop(['ID'], axis= 1)
df['diagnosis'] = (df['diagnosis'] == "M").astype(int)

'''''for i in df.columns:
    df[i] = (df[i] - min(df[i]))/(max(df[i] - min(df[i]))) '''
df.describe()
```

```python
x = df.drop('diagnosis', axis= 1)
y = df['diagnosis']

ros = RandomOverSampler()
x, y = ros.fit_resample(x, y)
```

```python
x = x.to_numpy()
y = y.to_numpy()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2)
```

```python
from collections import Counter
import numpy as np

def euclidean_dist(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))

def predict(x_train, x_test, y_train, y_test, k=3):
    labels = []
    for i in range(len(x_test)):
        # Calculate distances to ALL training samples
        distances = [euclidean_dist(x_test[i], x_train[j]) for j in range(len(x_train))]
        
        # Get k nearest neighbors
        k_indices = np.argsort(distances)[:k]
        k_nearest_labels = [y_train[idx] for idx in k_indices]
        
        # Majority vote
        label = Counter(k_nearest_labels).most_common()[0][0]
        labels.append(label)
    
    return labels  # Return the list, not single label

def accuracy(labels, y_test):
    count = np.sum(y_test == labels)  # Remove brackets
    acc = (count / len(y_test)) * 100
    return acc
```

```python
test_labels = predict(x_train, x_test, y_train, y_test, k= 9)
acc = accuracy(test_labels, y_test)
```
